## Problem 1

```python
# Plot inline
%matplotlib inline

import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
import scipy.signal as sg
import utilities.wavelet as wv
import utilities.image as im
import utilities.quantization as quant
import os
import subprocess
import copy

img = im.read_gecko_image()
org = copy.deepcopy(img)

h = sg.daub(2)

# Forward Transformation
wv.forward(h, img)
img_f = copy.deepcopy(img)

# Threshold + Encode
t, ltmax = quant.log_thresh(img)
img = quant.encode(img, t, ltmax)
img_e = copy.deepcopy(img)

# Store to file
filename = 'encoded_image'
img_e.tofile(filename)
file_size = os.stat(filename).st_size

# Compress File
subprocess.call(['gzip', filename])
c_file_size = os.stat(filename + '.gz').st_size

# Decompress File
subprocess.call(['gunzip', filename + '.gz'])
d_file_size = os.stat(filename).st_size

# Read from file
img = np.fromfile(filename).reshape(img.shape)

# Decode
img = quant.decode(img, t, ltmax)
img_d = copy.deepcopy(img)

wv.inverse(h, img)
img_i = copy.deepcopy(img)

plt.figure()
plt.imshow(org, cm.Greys_r)
plt.figure()
plt.imshow(img_f, cm.Greys_r)
plt.figure()
plt.imshow(img_e, cm.Greys_r)
plt.figure()
plt.imshow(img_d, cm.Greys_r)
plt.figure()
plt.imshow(img_i, cm.Greys_r)

print "Compression Level: %s" % (1 - float(c_file_size) / float(file_size))
```

Compute and Plot Symbol A(3)

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as sig
import utilities.wavelet as wv

h = sig.daub(3)
symbol = wv.compute_symbol(h)

f = np.linspace(0,1)
p = [np.absolute(symbol(v)) for v in f]

plt.figure()
plt.title('D6 Symbol')
plt.plot(f,p)
```

Compute and plot $\phi$ and $\psi$ (Numeric)

```python
%matplotlib inline 

import matplotlib.pyplot as plt
import scipy.signal as sig

h = sig.daub(3)
t, phi, psi = sig.wavelets.cascade(h)

plt.figure()
plt.title('D6 Scaling Function')
plt.plot(t,phi)
plt.figure()
plt.title('D6 Wavelet')
plt.plot(t,psi)
```
